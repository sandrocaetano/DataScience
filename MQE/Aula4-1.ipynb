```python
baseDados.sort()
serie = pd.Series(baseDados)

df = pd.DataFrame(serie.value_counts().sort_index())
df = df.reset_index()
df.columns = ['xi','fi']
total = df['fi'].sum()

freq_acumulada = []
acumulada = 0 
for frequencia in df['fi']:
    acumulada = acumulada + frequencia
    freq_acumulada.append(acumulada)

df['Fi'] = freq_acumulada
df['Fi %'] = df['Fi'] / total
df['xifi'] = df['xi'] * df['fi']
df['xi2fi'] = (df['xi'] ** 2) * df['fi']

df['filnxi'] = df['fi'] * np.log(df['xi'])
df['fi/xi'] = df['fi'] / df['xi']
df
```

```python
media = (1 / total) * df['xifi'].sum()
mediaGeometrica = np.e ** ((1/total) * df['filnxi'].sum())
mediaHarmonica = total / (df['fi/xi'].sum())
mediaQuadratica = np.sqrt(df['xi2fi'].sum()/total)


print('Media Aritmetica: ', np.round(media,4))
print('Media Geometrica: ', np.round(mediaGeometrica,4))
print('Media Harmonica: ', np.round(mediaHarmonica,4))
print('Media Quadratica: ', np.round(mediaQuadratica,4))
```
